```python
import numpy as np
a = np . array ([1 , 2 , 3])
b = np . array ([4 , 5 , 6])
dot_product = np . dot (a , b )
print ( dot_product )
```

```python
def cosine_similarity (x , y ) :
    return np . dot (x , y ) / ( np . linalg . norm ( x ) * np . linalg . norm ( y ) )
cos_sim = cosine_similarity (a , b )
print ( cos_sim )
```

```python
X = np . array ([[1 , 2] , [3 , 4] , [5 , 6]])
pairwise_dist = np . sqrt (
np . sum (( X [: , None , :] - X [ None , : , :]) **2 , axis = -1)
)
print ( pairwise_dist )
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

class1 = np.random.randn(100, 2) + [2, 2]
class2 = np.random.randn(100, 2) + [-2, -2]

plt.scatter(class1[:, 0], class1[:, 1], label='Class 1')
plt.scatter(class2[:, 0], class2[:, 1], label='Class 2')

plt.legend()
plt.show()
```
