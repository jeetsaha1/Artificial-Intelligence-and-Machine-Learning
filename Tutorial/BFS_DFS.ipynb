```python
def dfs_path(graph, start, goal):
    stack = [start]
    visited = set()
    parent = {start: None}

    while stack:
        node = stack.pop()

        if node in visited:
            continue

        visited.add(node)

        if node == goal:
            path = []
            while node is not None:
                path.append(node)
                node = parent[node]
            return path[::-1]   # reverse path

        for neighbor in graph[node]:
            if neighbor not in visited:
                parent[neighbor] = node
                stack.append(neighbor)

    return None



graph1 = {
    "1": ["2"],
    "2": ["1", "3", "4"],
    "3": ["2", "4", "7"],
    "4": ["2", "3", "5"],
    "5": ["4", "6"],
    "6": ["5", "7"],
    "7": ["3", "6"]
}

print(dfs_path(graph1, "1", "6"))
```






5 The Water Jug Problem

5.1 Problem Description

Two jugs of capacities 3L and 5L are given. Allowed operations:

• Fill a jug

• Empty a jug

• Pour water between jugs

Goal: Measure exactly 4 liters.

```python
from collections import deque

def water_jug_neighbours(state, cap1, cap2):
    a,b = state
    neighbours = []


    #Fill jugs
    neighbours.append((cap1,b))
    neighbours.append((a,cap2))

    # Empty jugs
    neighbours.append((0,b))
    neighbours.append((a,0))

    # Pour jug1 -> jug2
    pour = min(a,cap2-b)
    neighbours.append((a-pour ,b+pour))

    # Pour jug2 -> jug1
    pour = min(b,cap1-a)
    neighbours.append((a+pour,b-pour))

    return neighbours







def bfs_water_jug(cap1, cap2, target):
    start = (0,0)

    queue = deque([start])
    visited = {start}
    parent = {start : None}

    while queue:
        state = queue.popleft()
        a,b = state


        #base case
        if a == target or b == target:
            path = []
            while state is not None:
                path.append(state)
                state = parent[state]
            return path[::-1]

        for next_state in water_jug_neighbours(state,cap1, cap2):
            if next_state not in visited:
                visited.add(next_state)
                parent[next_state] = state
                queue.append(next_state)

    return None

print(bfs_water_jug(3,5,4))
```
