```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Reliance Industries Stock Price History.csv")

# Convert Date
df['Date'] = pd.to_datetime(df['Date'])

# Remove commas and convert to float
price_cols = ['Price', 'Open', 'High', 'Low']
for col in price_cols:
    df[col] = df[col].str.replace(',', '').astype(float)

# Remove % sign and convert
df['Change %'] = df['Change %'].str.replace('%', '').astype(float)

# Set Date as index
df.set_index('Date', inplace=True)

# Plot
df['Price'].plot(figsize=(12,6), title="Reliance Industries Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
from queue import Queue

graph1 = {
    "1": ["2"],
    "2": ["1", "3", "4"],
    "3": ["2", "4", "7"],
    "4": ["2", "3", "5"],
    "5": ["4", "6"],
    "6": ["5", "7"],
    "7": ["3", "6"]
}

q = Queue()
parent = {}
visited = {}
level = {}
BFS_Traversal_Output = []


source = "1"

#Initialisation
parent[source] = None
visited[source] =True
level[source] = 0


q.put(source)

while not q.empty():
    u = q.get()
    BFS_Traversal_Output.append(u)

    for neighbour in graph1[u]:
        if neighbour not in visited:
            visited[neighbour] = True
            parent[neighbour] = u
            level[neighbour] = level[u]+1
            q.put(neighbour)

print(BFS_Traversal_Output)

```
