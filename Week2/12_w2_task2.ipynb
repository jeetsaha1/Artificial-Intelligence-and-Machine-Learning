## 2Ô∏è‚É£ Apply a Suitable Algorithm to Locate the Scroll (Using DFS)

---

## üîπ Chosen Algorithm: **Depth-First Search (DFS)**

Depth-First Search (DFS) is a **suitable algorithm** for locating the **Ancient Scroll (S)** in the cursed library because:

- The goal state lies **deep in the graph**
- The graph contains **cycles**, which DFS can handle safely using a **visited set**
- DFS explores **one path fully** before considering alternative paths, making it effective for deep search spaces

---

## üîπ DFS Search Strategy

The DFS procedure follows these steps:

1. Start from the **initial state H (Hall)**
2. Expand the current node by moving **as deep as possible** along a single path
3. If a **dead end** or **previously visited node** is encountered, **backtrack**
4. Continue exploring unexplored branches until the **goal state S** is found

---

## üîπ DFS Traversal (High-Level Description)

- DFS initially follows the path:

$$
H \rightarrow A \rightarrow D \rightarrow E \rightarrow A
$$

- The **illusion corridor** causes a cycle, but the visited set prevents infinite looping
- DFS then **backtracks to H** and explores the next branch
- The search proceeds through the **B branch**
- After reaching the dead end:

$$
F \rightarrow X
$$

DFS backtracks again and continues via:

$$
B \rightarrow G \rightarrow I \rightarrow J \rightarrow S
$$

- The **goal state S** is eventually reached

---

## ‚úÖ Conclusion

Depth-First Search successfully locates the **Ancient Scroll** by fully exploring paths, handling cycles through backtracking, and systematically searching deeper branches until the goal is found.

## 5Ô∏è‚É£ Role of the Visited Set in Avoiding Infinite Loops (DFS)

The **visited set** plays a crucial role in **Depth-First Search (DFS)** when the graph contains **cycles**, as in the cursed library.

---

### üîπ Why the Visited Set Is Needed

The library contains an **illusion corridor** forming a cycle:

$$
A \rightarrow D \rightarrow E \rightarrow A
$$

Without a visited set, DFS would:

- Repeatedly revisit nodes **A**, **D**, and **E**
- Enter an **infinite loop**
- Fail to reach the **Ancient Scroll (S)**

---

### üîπ How the Visited Set Works in DFS

- Each node is **marked as visited** when it is first explored
- Before expanding a neighboring node, DFS checks:
  - If the node is **already visited ‚Üí it is skipped**
- This prevents DFS from re-entering cycles and **forces backtracking**

---

### ‚úÖ Result of Using a Visited Set

- Infinite traversal is **avoided**
- Illusion corridors are **safely bypassed**
- DFS continues exploring **unvisited paths**
- The search successfully progresses toward the **goal state**

---

## üìå Main Line

**The visited set prevents DFS from revisiting the same nodes repeatedly, thereby avoiding infinite loops caused by cycles in the graph.**

## 7Ô∏è‚É£ Comparison of DFS and BFS in the Cursed Library

---

###  Comparison Table

| Aspect | Depth-First Search (DFS) | Breadth-First Search (BFS) |
|------|--------------------------|----------------------------|
| **Search Strategy** | Explores **deeply first** | Explores **level by level** |
| **Handling Cycles** | Uses **visited set + backtracking** | Uses **visited set** |
| **Memory Usage** | **Low** (stores current path) | **High** (stores full frontier) |
| **Optimality** |  Not optimal |  Finds shortest path |
| **Behavior in Illusion Loop** | Gets trapped first, then backtracks | Avoids deep loop early |
| **Suitability Here** | Good for **deep goal**, memory-efficient | Good for **shortest path** |

---

## üîπ Behavioral Difference in This Library

### üî∏ DFS Behavior
- Initially explores the path:
  
$$
H \rightarrow A \rightarrow D \rightarrow E
$$

- Encounters the **illusion loop**
- **Backtracks** after detecting visited nodes
- Eventually explores the **B branch** and reaches the goal **S**

---

### üî∏ BFS Behavior
- Explores **all neighbors of H first**
- Reaches the path:

$$
H \rightarrow B \rightarrow G
$$

- Finds the scroll **earlier with fewer expansions**
- Guarantees the **shortest path**

---

## üß† Key Insight

- **BFS** is **optimal** and guarantees the shortest path.
- **DFS** is **memory-efficient** and sufficient when the goal exists.
- Since the **shortest path is not required**, DFS is an acceptable choice.

---

## üìåConclusion Line

**While BFS guarantees the shortest path, DFS is more memory-efficient and successfully locates the scroll through backtracking, making it suitable for this problem despite non-optimality.**


```python
# DFS implementation for the Cursed Library problem

# Graph representation (Adjacency List)
graph = {
    "H": ["A", "B", "C"],
    "A": ["D"],
    "D": ["E"],
    "E": ["A"],      # illusion loop
    "B": ["F", "G"],
    "F": ["X"],      # dead end
    "X": [],
    "G": ["I"],
    "I": ["J"],
    "J": ["S"],      # goal
    "C": ["Y"],      # dead end
    "Y": [],
    "S": []
}

# DFS function
def dfs(current, goal, visited, path):
    # Mark current node as visited
    visited.add(current)
    path.append(current)

    # Check if goal is reached
    if current == goal:
        return True

    # Explore neighbors
    for neighbor in graph[current]:
        if neighbor not in visited:
            if dfs(neighbor, goal, visited, path):
                return True

    # Backtracking step
    path.pop()
    return False


# Driver code
visited = set()
path = []

start = "H"
goal = "S"

found = dfs(start, goal, visited, path)

if found:
    print("Scroll found!")
    print("Path:", " -> ".join(path))
else:
    print("Scroll not found")
```
