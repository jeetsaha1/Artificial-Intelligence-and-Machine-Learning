# Problem 1: The Museum Night Patrol

A national museum deploys a **night guard** who must patrol through interconnected rooms using **one-way corridors**.  
Each corridor traversal takes **exactly one unit of time**.

---

## üèõÔ∏è Museum Layout (Directed Graph)

The museum layout is represented as a **directed graph** with the following corridors:

- **Entrance (E)** ‚Üí Rooms **A**, **B**
- **Room A** ‚Üí Rooms **C**, **D**
- **Room B** ‚Üí Room **D**
- **Room C** ‚Üí Room **F**
- **Room D** ‚Üí Rooms **C**, **G**
- **Room F** ‚Üí Room **D** *(forming a loop)*
- **Room G** ‚Üí **Vault (V)**

### ‚úÖ Chosen Algorithm: **Breadth-First Search (BFS)**

- **Shallow Goal State**

### ‚úÖ Conclusion

Breadth-First Search is **complete**, **optimal**, and **safe** for this problem.  
It guarantees finding the **shortest valid path** from **Entrance (E)** to **Vault (V)** while respecting all constraints.



## 3Ô∏è‚É£ Clearly Show the Order of Node Expansion

Breadth-First Search (BFS) expands nodes **level by level** using a **FIFO queue** and a **visited set**.  
When multiple successors exist at the same depth, the **successor ordering** determines the expansion order.

In this case, successors are considered in **alphabetical order**.

---

## üß† BFS Node Expansion (Step-by-Step)

## 5.  Prove whether BFS guarantees the optimal solution under the given constraints.

---

### üîπ Claim

**Breadth-First Search (BFS) guarantees an optimal solution** for the *Museum Night Patrol* problem.

---

### üîπ Proof

- Each corridor traversal has a **uniform cost of 1 unit time**.
- Therefore, the problem is an **unweighted state-space graph**.

BFS explores the state space **level by level**, where:

- **Level 0** ‚Üí Path cost = 0  
- **Level 1** ‚Üí Path cost = 1  
- **Level 2** ‚Üí Path cost = 2  
- **Level 3** ‚Üí Path cost = 3  
- **Level 4** ‚Üí Path cost = 4  

- BFS uses a **FIFO (First-In, First-Out) queue**, ensuring that **all nodes at depth _d_ are expanded before any node at depth _d + 1_**.
- Hence, when the goal state **V** is first reached at **depth 4**, **all possible paths with cost less than 4 have already been explored**.
- Therefore, **no shorter (lower-cost) path to the goal can exist**.

---

### üîπ Conclusion

Since BFS always finds the goal at the **minimum depth**, and **all edge costs are equal**, BFS is **both complete and optimal** under the given constraints.



## 6. Analyze the time and space complexity of the algorithm.

Let:

- **b** = branching factor (maximum number of successors of a node)
- **d** = depth of the shallowest solution  
  (here, **d = 4**)

---

## ‚è±Ô∏è Time Complexity

- Breadth-First Search explores **all nodes up to depth _d_**.
- The total number of nodes expanded is:

$$
T(b, d) = O(b^d)
$$

### For the given problem:
- Maximum branching factor:  
  $$ b \leq 2 $$
- Depth of solution:  
  $$ d = 4 $$

Therefore,

$$
T = O(2^4) = O(16)
$$

This indicates that BFS explores a **small and manageable number of states**.

---

## üíæ Space Complexity

- BFS stores:
  - All nodes in the **frontier (queue)**
  - All **visited states**

Thus, the space complexity is:

$$
S(b, d) = O(b^d)
$$

For this problem:

$$
S = O(2^4) = O(16)
$$

---

## ‚úÖ Conclusion

Although BFS has **exponential time and space complexity**, it is **acceptable in this problem** due to the **small branching factor** and **strict depth limit**.


## 7. Explain what would happen if the visited set were removed.

---

### üö´ Without a Visited Set

The state-space graph contains a **cycle**:

$$
D \rightarrow C \rightarrow F \rightarrow D
$$

Without maintaining a visited set, BFS would **revisit the same states repeatedly**.

This leads to the following problems:

- **Infinite or redundant expansions** of states  
- **Exponential growth** in the number of nodes explored  
- **Increased time and memory consumption**  
- **Possible non-termination** in cyclic graphs  

---

### ‚ö†Ô∏è Example of Repeated Expansion

Without visited-state tracking, BFS may repeatedly follow the cycle:

$$
D \rightarrow C \rightarrow F \rightarrow D \rightarrow C \rightarrow F \rightarrow \cdots
$$

This loop consumes available moves and **prevents efficient discovery of the goal state**.

---

### ‚úÖ With a Visited Set

- Each state is **expanded only once**
- Cycles are **safely avoided**
- Redundant paths are eliminated
- BFS remains **complete**, **optimal**, and **efficient**

---

## ‚úÖ Conclusion

The visited set is **essential** for BFS in graphs containing cycles.  
Without it, BFS may become inefficient or fail to terminate, whereas with it, BFS guarantees correct and optimal performance.


## **Here is the program for the above problem**

```python
from collections import deque

# Directed graph
graph = {
    'E': ['A', 'B'],
    'A': ['C', 'D'],
    'B': ['D'],
    'C': ['F'],
    'D': ['C', 'G'],
    'F': ['D'],
    'G': ['V'],
    'V': []
}

start = 'E'
goal = 'V'
max_depth = 4

# BFS queue: (current_node, path, depth)
queue = deque()
queue.append((start, [start], 0))

visited = set()
visited.add(start)

found = False

while queue:
    node, path, depth = queue.popleft()

    if node == goal:
        print("The roadmap to reach the vault :", " -> ".join(path))
        found = True
        break

    if depth < max_depth:
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor], depth + 1))

if not found:
    print("No valid path within 4 moves")
```
