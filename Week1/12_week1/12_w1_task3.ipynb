```python
import numpy as np
import matplotlib.pyplot as plt

# Fix random seed for reproducibility
np.random.seed(123)
```

```python
# Generate synthetic dataset
X = np.random.randn(100, 2)

# Generate labels (simple rule-based classification)
y = (X[:, 0] + X[:, 1] > 0).astype(int)

print("First 5 data points:\n", X[:5])
print("First 5 labels:", y[:5])
```

```python
# Mock predictions (simulating a trained model)
predictions = y.copy()

print("First 5 predictions:", predictions[:5])
```

```python
# Save experiment results
np.save("predictions.npy", predictions)

print("Predictions saved successfully.")
```

```python
# Load saved results
loaded_predictions = np.load("predictions.npy")

print("Predictions loaded successfully.")
```

```python
# Verify if results are identical
print("Results identical:",
      np.array_equal(predictions, loaded_predictions))
```
