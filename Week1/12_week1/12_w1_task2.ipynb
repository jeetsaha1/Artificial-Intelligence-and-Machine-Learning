```python
# 1. Generate a synthetic dataset with two linearly separable classes.

# numpy is used to generate the synthetic data
import numpy as np
# matplotlib is used to plot and visualize dataset.
import matplotlib.pyplot as plt

# Fixing random seed for reproducibility.
np.random.seed(42)

# Generate two well-separated random values
class1_linear = np.random.randn(100, 2) + [2, 2]
class2_linear = np.random.randn(100, 2) + [-2, -2]

# Plot linearly separable data
plt.figure()
plt.scatter(class1_linear[:, 0], class1_linear[:, 1], label='Class 1')
plt.scatter(class2_linear[:, 0], class2_linear[:, 1], label='Class 2')
plt.title("Linearly Separable Dataset")
plt.legend()
plt.show()
```

```python
# 2. Generate another dataset where the classes are not linearly separable.

# Generates 50 angles between 0 and 2Ï€.
theta = np.linspace(0, 2*np.pi, 50)

# Inner and outer circles for two synthetic datasets
class1_nonlinear = np.c_[np.cos(theta), np.sin(theta)]
class2_nonlinear = 2 * class1_nonlinear

# Plot non-linearly separable data
plt.figure()
plt.scatter(class1_nonlinear[:, 0], class1_nonlinear[:, 1], label='Class 1')
plt.scatter(class2_nonlinear[:, 0], class2_nonlinear[:, 1], label='Class 2')
plt.title("Non-Linearly Separable Dataset")
plt.legend()
plt.show()
```
